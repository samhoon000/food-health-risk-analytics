```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("../Data/nutrients/FOOD-DATA-GROUP1.csv")
df2 = pd.read_csv("../Data/nutrients/FOOD-DATA-GROUP2.csv")
df3 = pd.read_csv("../Data/nutrients/FOOD-DATA-GROUP3.csv")
df4 = pd.read_csv("../Data/nutrients/FOOD-DATA-GROUP4.csv")
df5 = pd.read_csv("../Data/nutrients/FOOD-DATA-GROUP5.csv")
```

```python
df=pd.concat([df1,df2,df3,df4,df5],ignore_index=True)
```

```python
df=df.drop(columns=['Unnamed: 0.1', 'Unnamed: 0'])
```

```python
df.columns=(
    df.columns
    .str.strip()
    .str.lower()
    .str.replace(' ','_')
)
```

```python
# Outliers
df[
    df["caloric_value"] > 1000
][["food", "caloric_value"]]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df["caloric_value"], bins=50)
plt.xlabel("Calories")
plt.ylabel("Count")
plt.title("Caloric Value Distribution")
plt.show()
```

```python
Q1=df.caloric_value.quantile(0.25)
Q3=df.caloric_value.quantile(0.75)
IQR=Q3-Q1
lower_limit=Q1-1.5*IQR
upper_limit=Q3+1.5*IQR
```

```python
print(lower_limit)
print(upper_limit)
```

```python
original_rows=len(df)
```

```python
df = df[
    (df["caloric_value"] >= 0) &
    (df["caloric_value"] <= upper_limit)
]
```

```python
original_rows-len(df)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.hist(df["caloric_value"],bins=50)

plt.xlabel("Caloric Value")
plt.ylabel("Count")
plt.title("Caloric Value Distribution After Outlier Removal")
plt.show()
```
