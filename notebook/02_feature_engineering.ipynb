```python
import pandas as pd
df = pd.read_csv("../Data/Cleaned/Cleaned_df1.csv")
```

```python
df["protein_calorie_ratio"] = (
    df["protein"] /
    (df["caloric_value"] +1)
)
```

```python
df['protein_calorie_ratio'].describe()
```

```python
#Extreme protein-to-calorie ratios caused by inconsistent nutrient measurements were filtered to improve feature reliability.
df = df[
    df["protein_calorie_ratio"] <= 0.5
]
```

```python
df['protein_calorie_ratio'].describe()
```

```python
df["health_risk_percentage"] = (
    (
        (df["sugars"] / 50) * 35 +
        (df["saturated_fats"] / 20) * 30 +
        (df["sodium"] / 2300) * 20 +
        (df["cholesterol"] / 300) * 5 -
        (df["dietary_fiber"] / 30) * 10
    )
).clip(0, 100)
```

```python
df['health_risk_percentage'].describe()
```
