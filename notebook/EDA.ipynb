```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
food_nutrition = pd.read_csv("../Data/food_nutritions/food_nutrient.csv")
food = pd.read_csv("../Data/food_nutritions/food.csv")
food_category = pd.read_csv("../Data/food_nutritions/food_category.csv")
nutrient = pd.read_csv("../Data/food_nutritions/nutrient.csv")
food_portion= pd.read_csv("../Data/food_nutritions/food_portion.csv")
```

```python
nutrition_df=food_nutrition.merge(
    nutrient,
    left_on='nutrient_id',
    right_on='id',
    how='left'
)
```

```python
nutrition_df.head()
```

```python
df2=nutrition_df.merge(
    food,
    on='fdc_id',
    how='left'
)
```

```python
df2.head()
```

```python
df2['name'].unique()
```

```python
# Select only the important nutrients that actually affect health
important_nutrients = [
    "Protein",
    "Total lipid (fat)",
    "Carbohydrate, by difference",
    "Total Sugars",
    "Fiber, total dietary",
    "Sodium, Na",
    "Potassium, K",
    "Cholesterol",
    "Energy",
    "Vitamin C, total ascorbic acid",
    "Calcium, Ca",
    "Iron, Fe"
]

df_nutrients=df2[df2['name'].isin(important_nutrients)]
```

```python
df = df_nutrients[
    [
        "description",
        "name",
        "amount",
        "food_category_id"
    ]
]
```

```python
df.head()
```

```python
final_df=df.pivot_table(
    index='description',
    columns='name',
    values='amount',
    aggfunc='mean'
).reset_index()
```
